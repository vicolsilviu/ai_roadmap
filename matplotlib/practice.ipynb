```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create plot with customizations
plt.plot(x, y, color='green', linestyle='--', marker='o', markerfacecolor='blue', markersize=10)

# Add title and labels
plt.title("Customized Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a new figure
fig = plt.figure()

# Add a set of axes to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot data
ax.plot([1, 2, 3, 4, 5], [2, 3, 5, 7, 11])

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot data in each subplot
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 1].plot([1, 2, 3], [1, 2, 3])
axs[1, 0].plot([1, 2, 3], [1, 0, 1])
axs[1, 1].plot([1, 2, 3], [3, 2, 1])

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure with custom size and DPI
fig = plt.figure(figsize=(8, 6), dpi=100)

# Add a set of axes to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot data
ax.plot([1, 2, 3, 4, 5], [2, 3, 5, 7, 11])

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Show plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create plot
plt.plot(x, y)

# Annotate a point
plt.annotate('Peak', xy=(5, 11), xytext=(3, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show plot
plt.show()
```
